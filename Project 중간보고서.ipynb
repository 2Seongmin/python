```python
station = subway.역명
station.value_counts()

# 서울에 위치한 역의 개수가 244개 (csv파일상)
```

```python
# 하루동안의 승객수 구해서 '하루승객수'열 만들기
cols =['06시이전', '06-07시간대','07-08시간대','08-09시간대','09-10시간대','10-11시간대','11-12시간대',
       '12-13시간대','13-14시간대','14-15시간대','15-16시간대','16-17시간대','17-18시간대','18-19시간대',
       '19-20시간대','20-21시간대','21-22시간대','22-23시간대','23-24시간대','24시이후']
subway['하루승객수'] = subway.loc[:,cols].sum(axis=1) 
# 'cols에 해당하는 열의 데이터만 추출 후, 열 방향으로 합을 구함
subway.하루승객수
```

```python
# 2022년동안 일별 하루승객수가 가장 많았던 역명
(subway[["호선","역명","수송일자","승하차구분","하루승객수"]]
 .sort_values(["수송일자","하루승객수"],ascending=[True,False])
 .drop_duplicates(subset="수송일자"))
```

```python
# 수송일자의 시작이 몇번째 행부터 시작되는 지 파악할 수 있음
(subway[["연번","호선","역명","수송일자","승하차구분","하루승객수"]]
 .sort_values(["연번"],ascending=True)
 .drop_duplicates(subset="수송일자"))
```

```python
# 원하는 날짜의 시작과 다음 날짜의 시작 전까지 범위를 설정해서 값을 새로 생성
# 만약 수송일자가 2022-01-01 일때, 승객 수가 많았던 역 순위

festival_date = subway.iloc[0:544]
festival_date
```

```python
# festival_date(2022-01-01)에 가장 많은 승객이 이용한 전철역
(festival_date[["호선","역명","수송일자","승하차구분","하루승객수"]]
 .sort_values(["하루승객수"],ascending=False))
```
